## **CREDIT CARD FRAUD DETECTION**

**IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**LOADING AND PREPROCESSING DATASET**

```python
df=pd.read_csv("D:\codsoft\Credit Card Fraud Detection\creditcard.csv")
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

UNDERSAMPLING - making data ready for training

```python
normal=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
normal_sample=normal.sample(n=492)
```

```python
new_df=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_df.head(5)
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

**SPLITTING THE DATASET**

```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

**MODEL TRAINING**

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

**ACCURACY SCORE USING METRICS**

```python
X_train_prediction=model.predict(X_train)
training_data_acuracy=accuracy_score(X_train_prediction,Y_train)*100
```

```python
print(f"Training Data Accuracy: {training_data_acuracy}%")
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)*100
```

```python
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
